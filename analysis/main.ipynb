### Data Analysis and Visualizations

```python
canceled_percentage = df["is_canceled"].value_counts(normalize=True) # return value: %
print(canceled_percentage)
```

- Non-Cancelation ~ 62.86%
- Cancelation ~ 37.13%

```python
plt.figure(figsize=(5, 3))
plt.title("Reservation Count")
plt.bar(['Not canceled', 'Canceled'], df['is_canceled'].value_counts(), edgecolor = 'k', width=0.7)
plt.savefig("C:\\Users\\soume\\OneDrive - sk2002\\Data Analysis Report - Hotel Booking\\outputs\\reservation_count", dpi=800)
plt.show()
```

- Rate of cancellation is greater than half of 'Not Cancelled'

```python
plt.figure(figsize=(10, 4))
ax1 = sns.countplot(x = 'hotel', hue = 'is_canceled', data=df, palette='Blues')
plt.title("Reservation status (Resort Hotel VS City Hotel)", size = 20)
plt.xlabel("hotel")
plt.ylabel("number of reservations")
plt.legend(["not canceled", "canceled"])
plt.savefig("C:\\Users\\soume\\OneDrive - sk2002\\Data Analysis Report - Hotel Booking\\outputs\\cancelation rate in hotels", dpi=1200)
plt.show()
```

- Cancelation in Resort Hotel is much less than as of City Hotel

```python
resort_hotel = df[df['hotel'] == 'Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize=True)
```

```python
city_hotel = df[df['hotel'] == 'City Hotel']
city_hotel['is_canceled'].value_counts(normalize=True)
```

#### Resort Hotel
- Non-Canceled ~ 72%
- Canceled ~ 28%

#### City Hotel
- Non-Canceled ~ 58%
- Canceled ~ 41.7%

```python
# in order to filter cancelation in resort_hotel - reservation_status_date
# group_by
# taking adr of one day as mean value
resort_hotel = resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel= city_hotel.groupby('reservation_status_date')[['adr']].mean()

plt.figure(figsize=(20, 8))
plt.plot(resort_hotel.index, resort_hotel['adr'], label = "resort hotel")
plt.plot(city_hotel.index, city_hotel['adr'], label = "city hotel")
plt.title("Average Daily Rate (Resort Hotel VS City Hotel)")
plt.legend(fontsize = 20)
plt.savefig("C:\\Users\\soume\\OneDrive - sk2002\\Data Analysis Report - Hotel Booking\\outputs\\adr_in_hotels", dpi=800)
plt.show()
```

- __City Hotel__ price for reservation is lower than __Resort Hotel__ in some days

```python
df['month'] = df['reservation_status_date'].dt.month # for only month
plt.figure(figsize=(16, 8))
ax1 = sns.countplot(x = 'month', hue = 'is_canceled', data=df)
plt.title("Reservation status per month", size=30)
plt.xlabel("month")
plt.ylabel("number of reservations")
plt.legend(['not canceled', 'canceled'])
plt.savefig("C:\\Users\\soume\\OneDrive - sk2002\\Data Analysis Report - Hotel Booking\\outputs\\reservation_status_per_month", dpi=1200)
plt.show()
```

- Non-Cancelation rate is having a peak during the month of July & August
- Cancelation rate is having a peak during the month of January, July and October

- Non-Cancelation rate is low in the month of December
- Cancelation rate is lower during the months of July & August
