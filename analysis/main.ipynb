```python
plt.figure(figsize=(10, 4))
ax1 = sns.countplot(x = 'hotel', hue = 'is_canceled', data=df, palette='Blues')
plt.title("Reservation status (Resort Hotel VS City Hotel)", size = 20)
plt.xlabel("hotel")
plt.ylabel("number of reservations")
plt.legend(["not canceled", "canceled"])
plt.savefig("C:\\Users\\soume\\OneDrive - sk2002\\Data Analysis Report - Hotel Booking\\outputs\\cancelation rate in hotels", dpi=1200)
plt.show()
```

- Cancelation in Resort Hotel is much less than as of City Hotel

```python
resort_hotel = df[df['hotel'] == 'Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize=True)
```

```python
city_hotel = df[df['hotel'] == 'City Hotel']
city_hotel['is_canceled'].value_counts(normalize=True)
```

#### Resort Hotel
- Non-Canceled ~ 72%
- Canceled ~ 28%

#### City Hotel
- Non-Canceled ~ 58%
- Canceled ~ 41.7%

```python
# in order to filter cancelation in resort_hotel - reservation_status_date
# group_by
# taking adr of one day as mean value
resort_hotel = resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel= city_hotel.groupby('reservation_status_date')[['adr']].mean()

plt.figure(figsize=(20, 8))
plt.plot(resort_hotel.index, resort_hotel['adr'], label = "resort hotel")
plt.plot(city_hotel.index, city_hotel['adr'], label = "city hotel")
plt.title("Average Daily Rate (Resort Hotel VS City Hotel)")
plt.legend(fontsize = 20)
plt.savefig("C:\\Users\\soume\\OneDrive - sk2002\\Data Analysis Report - Hotel Booking\\outputs\\adr_in_hotels", dpi=800)
plt.show()
```

- __City Hotel__ price for reservation is lower than __Resort Hotel__ in some days

```python
df['month'] = df['reservation_status_date'].dt.month # for only month
plt.figure(figsize=(16, 8))
ax1 = sns.countplot(x = 'month', hue = 'is_canceled', data=df)
plt.title("Reservation status per month", size=30)
plt.xlabel("month")
plt.ylabel("number of reservations")
plt.legend(['not canceled', 'canceled'])
plt.savefig("C:\\Users\\soume\\OneDrive - sk2002\\Data Analysis Report - Hotel Booking\\outputs\\reservation_status_per_month", dpi=1200)
plt.show()
```

- Non-Cancelation rate is having a peak during the month of July & August
- Cancelation rate is having a peak during the month of January, July and October

- Non-Cancelation rate is low in the month of December
- Cancelation rate is lower during the months of July & August

```python
# inorder to check if reservation price has effect on cancelation rate
plt.figure(figsize=(15, 8))
plt.title("ADR per month", fontsize=30)

# filter the dataframe and group by month, summing up 'adr'
sns.barplot(x ='month', y ='adr', data=df[df['is_canceled'] == 1].groupby('month')[['adr']].sum().reset_index())
plt.show()
```

- Higher Price ↗️ More Cancelation Rate ↗️

```python
canceled_data_adr = canceled_data.groupby(
    'reservation_status_date')[['adr']].mean()
canceled_data_adr.reset_index(inplace=True)
canceled_data_adr.sort_values('reservation_status_date', inplace=True)

not_canceled_data = df[df['is_canceled'] == 0]
not_canceled_data_adr = not_canceled_data.groupby(
    'reservation_status_date')[['adr']].mean()
not_canceled_data_adr.reset_index(inplace=True)
not_canceled_data_adr.sort_values('reservation_status_date', inplace=True)

plt.figure(figsize=(20, 6))
plt.title("Average Daily Rate (ADR)")
plt.plot(not_canceled_data_adr['reservation_status_date'],
         not_canceled_data_adr['adr'], label='not canceled')
plt.plot(canceled_data_adr['reservation_status_date'],
         canceled_data_adr['adr'], label='canceled')
plt.legend()
plt.savefig("C:\\Users\\soume\\OneDrive - sk2002\\Data Analysis Report - Hotel Booking\\outputs\\adr_data", dpi=800)
plt.show()
```

#### Let us remove the inconsistency in data

```python
canceled_data_adr = canceled_data_adr[(canceled_data_adr['reservation_status_date'] > '2016') & (
    canceled_data_adr['reservation_status_date'] < '2017-09')]

not_canceled_data_adr = not_canceled_data_adr[(not_canceled_data_adr['reservation_status_date'] > '2016') & (
    not_canceled_data_adr['reservation_status_date'] < '2017-09')]
```

```python
plt.figure(figsize=(20, 6))
plt.title("Average Daily Rate (ADR)", fontsize=30)
plt.plot(not_canceled_data_adr['reservation_status_date'], not_canceled_data_adr['adr'], label = 'not canceled')
plt.plot(canceled_data_adr['reservation_status_date'], canceled_data_adr['adr'], label = 'canceled_data')
plt.legend(fontsize = 20)
plt.show()
```

- __Average Daily Rate__ is one of the most influential factor affecting cancellation rate
